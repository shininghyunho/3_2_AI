 첫 번째 단계의 합성곱 필터와 풀링 계층을 정의한다.
- 1개의 28*28 픽셀 이미지가 64개의 필터와 풀링층을 거쳐서(크기줄임)
- 14*14 픽셀 64개의 이미지가 나옴

```python
num_filters1 = 64

x = tf.placeholder(tf.float32, [None, 784])
x_image = tf.reshape(x, [-1,28,28,1])
#x_image = tf.image.central_crop(image = x_image, central_fraction = 0.9)
#x_image=tf.image.flip_left_right(x_image)
#x_image=tf.image.random_flip_left_right(
#    x_image,
#    seed=0.5
#)

W_conv1 = tf.Variable(tf.truncated_normal([3,3,1,num_filters1],
                                          stddev=0.1))
# 컨볼루션 레이어
h_conv1 = tf.nn.conv2d(x_image, W_conv1,
                       strides=[1,1,1,1], padding='SAME')

b_conv1 = tf.Variable(tf.constant(0.1, shape=[num_filters1]))
h_conv1_cutoff = tf.nn.relu(h_conv1 + b_conv1)

h_pool1 = tf.nn.max_pool(h_conv1_cutoff, ksize=[1,2,2,1],
                         strides=[1,2,2,1], padding='SAME')
```

 두 번째 단계의 합성곱 필터와 풀링 계층을 정의한다.
- 64개의 14*14 픽셀 이미지가 64개의 필터와 풀링층을 거쳐서(크기 줄임)
- 7*7 픽셀 64개의 이미지가 나옴

```python
num_filters2 = 64

W_conv2 = tf.Variable(
            tf.truncated_normal([3,3,num_filters1,num_filters2],
                                stddev=0.1))
h_conv2 = tf.nn.conv2d(h_pool1, W_conv2,
                       strides=[1,1,1,1], padding='SAME')

b_conv2 = tf.Variable(tf.constant(0.1, shape=[num_filters2]))
h_conv2_cutoff = tf.nn.relu(h_conv2 + b_conv2)

h_pool2 = tf.nn.max_pool(h_conv2_cutoff, ksize=[1,2,2,1],
                         strides=[1,2,2,1], padding='SAME')
```

 세 번째 단계의 합성곱 필터와 풀링 계층을 정의한다.
- 64개의 7*7 픽셀 이미지가 64개의 필터와 풀링층을 거쳐서(크기 안줄임)
- 7*7 픽셀 64개의 이미지가 나옴

```python
num_filters3 = 64

W_conv3 = tf.Variable(
            tf.truncated_normal([3,3,num_filters2,num_filters3],
                                stddev=0.1))
h_conv3 = tf.nn.conv2d(h_pool2, W_conv3,
                       strides=[1,1,1,1], padding='SAME')

b_conv3 = tf.Variable(tf.constant(0.1, shape=[num_filters3]))
h_conv3_cutoff = tf.nn.relu(h_conv3 + b_conv3)

h_pool3 = tf.nn.max_pool(h_conv3_cutoff, ksize=[1,1,1,1],
                         strides=[1,1,1,1], padding='SAME')
```

네 번째 단계의 합성곱 필터와 풀링 계층을 정의한다.

- 64개의 7*7 픽셀 이미지가 64개의 필터와 풀링층을 거쳐서(크기 안줄임)
- 7*7 픽셀 64개의 이미지가 나옴

```python
num_filters4 = 64

W_conv4 = tf.Variable(
            tf.truncated_normal([3,3,num_filters3,num_filters4],
                                stddev=0.1))
h_conv4 = tf.nn.conv2d(h_pool3, W_conv4,
                       strides=[1,1,1,1], padding='SAME')

b_conv4 = tf.Variable(tf.constant(0.1, shape=[num_filters4]))
h_conv4_cutoff = tf.nn.relu(h_conv4 + b_conv4)

h_pool4 = tf.nn.max_pool(h_conv4_cutoff, ksize=[1,1,1,1],
                         strides=[1,1,1,1], padding='SAME')
```

전 결합층, 드롭아웃 계층, 소프트맥스 함수를 정의한다.
- 전결합층 : 마지막 합성곱 풀링 계층의 출력을 입력으로(num_units1)
- 전결합층 노드수 : 일종의 히든 레이어(num_units2)


```python
h_pool1_flat = tf.reshape(h_pool4, [-1, 7*7*num_filters4])

num_units1 = 7*7*num_filters4
num_units2 = 512 #2배로 늘려봄

# 전결합층

w2 = tf.Variable(tf.truncated_normal([num_units1, num_units2]))
b2 = tf.Variable(tf.constant(0.1, shape=[num_units2]))
hidden2 = tf.nn.relu(tf.matmul(h_pool1_flat, w2) + b2)

# 드랍아웃

keep_prob = tf.placeholder(tf.float32)
hidden2_drop = tf.nn.dropout(hidden2, keep_prob)

# 소프트맥스 (10개의 입력으로 맞춤)

w0 = tf.Variable(tf.zeros([num_units2, 10]))
b0 = tf.Variable(tf.zeros([10]))
p = tf.nn.softmax(tf.matmul(hidden2_drop, w0) + b0)
```

 오차 함수 loss, 트레이닝 알고리즘 train_step, 정답률 accuracy을 정의한다.

```python
t = tf.placeholder(tf.float32, [None, 10])
loss = -tf.reduce_sum(t * tf.log(p))
train_step = tf.train.AdamOptimizer(0.0001).minimize(loss)
correct_prediction = tf.equal(tf.argmax(p, 1), tf.argmax(t, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

 세션을 준비하고 Variable을 초기화한다.

```python
sess = tf.InteractiveSession()
sess.run(tf.initialize_all_variables())
saver = tf.train.Saver()
```

## 파라미터 최적화를 20000회 반복한다.

최종적으로 테스트 세트에 대해 약 99%의 정답률을 얻을 수 있다.

```python
i = 0
for _ in range(20000):
    i += 1
    batch_xs, batch_ts = mnist.train.next_batch(50)
    sess.run(train_step,
             feed_dict={x:batch_xs, t:batch_ts, keep_prob:0.7})
    if i % 2000 == 0:
        loss_vals, acc_vals = [], []
        for c in range(4):
            start = len(mnist.test.labels) / 4 * c
            end = len(mnist.test.labels) / 4 * (c+1)
            loss_val, acc_val = sess.run([loss, accuracy],
                feed_dict={x:mnist.test.images[start:end],
                           t:mnist.test.labels[start:end],
                           keep_prob:1.0})
            loss_vals.append(loss_val)
            acc_vals.append(acc_val)
        loss_val = np.sum(loss_vals)
        acc_val = np.mean(acc_vals)
        print ('Step: %d, Loss: %f, Accuracy: %f'
               % (i, loss_val, acc_val))
        saver.save(sess, './drive/My Drive/ai/cnn_session', global_step=i)
```
